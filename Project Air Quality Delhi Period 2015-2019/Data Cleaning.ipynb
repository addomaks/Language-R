```python
df['PM10_AVG24H'] = round(df['PM10'].rolling(window= 24, min_periods=2).mean(), 2)

df['PM2.5_AVG24H'] = round(df['PM2.5'].rolling(window=24, min_periods=2).mean(), 2)

df['SO2_AVG24H'] = round(df['SO2'].rolling(window=24, min_periods=2).mean(), 2)

df['CO_AVG8H'] = round(df['CO'].rolling(window=8, min_periods=1).mean(), 2)

df['Ozone_AVG8H'] = round(df['Ozone'].rolling(window=8, min_periods=1).mean(), 2)

df['NO2_AVG24H'] = round(df['NO2'].rolling(window=24, min_periods=2).mean(), 2)
```

```python
def SubIndex_PM10(Ip):
    if Ip <= 50:
        return 0 + (Ip - 0) * ((50 - 0) / (30 - 0))
    elif Ip <= 100:
        return 51 + (Ip - 51) * ((100-51) / (100-51))
    elif Ip <= 250:
        return 101 + (Ip - 101) * ((200-101) / (250-101))
    elif Ip <= 350:
        return 251 + (Ip - 251)
    elif Ip <= 430:
        return 301 + (Ip - 351) * (99/79)
    elif Ip >= 431:
        return 401 + (Ip - 431) * (598/431)
    else:
        return 0
    
df['PM10_SubIndex'] = df['PM10_AVG24H'].apply(lambda Ip: SubIndex_PM10(Ip)).astype('int')
```

```python
def SubIndex_PM25(Ip):
    if Ip <= 50:
        return 0 + (Ip - 0) * ((50 - 0) / (30 - 0))
    elif Ip <= 60:
        return 51 + (Ip - 31) * ((100-51) / (60-31))
    elif Ip <= 90:
        return 101 + (Ip - 61) * ((200-101) / (90-61))
    elif Ip <= 120:
        return 201 + (Ip - 91) * ((300-201)/(120-91))
    elif Ip <= 250:
        return 301 + (Ip - 121) * ((400-301)/(250-121))
    elif Ip >= 251:
        return 401 + (Ip - 431) * (598/251)
    else:
        return 0
    
df['PM2.5_SubIndex'] = df['PM2.5_AVG24H'].apply(lambda Ip: SubIndex_PM25(Ip)).astype('int')
```

```python
def SubIndex_SO2(Ip):
    if Ip <= 40:
        return 0 + (Ip - 0) * ((50 - 0) / (40 - 0))
    elif Ip <= 80:
        return 51 + (Ip - 41) * ((100-51) / (80-41))
    elif Ip <= 380:
        return 101 + (Ip - 81) * ((200-101) / (380-81))
    elif Ip <= 800:
        return 201 + (Ip - 381) * ((300-201)/(800-381))
    elif Ip <= 1600:
        return 301 + (Ip - 801) * ((400-301)/(1600-801))
    elif Ip >= 1601:
        return 401 + (Ip - 1601) * (598/1601)
    else:
        return 0
    
df['SO2_SubIndex'] = df['SO2_AVG24H'].apply(lambda Ip: SubIndex_SO2(Ip)).astype('int')
```

```python
def SubIndex_CO(Ip):
    if Ip <= 1.0:
        return 0 + (Ip - 0) * ((50 - 0) / (1.0 - 0))
    elif Ip <= 2.0:
        return 51 + (Ip - 1.1) * ((100-51) / (1.1-2.0))
    elif Ip <= 10.0:
        return 101 + (Ip - 2.1) * ((200-101) / (10.0-2.1))
    elif Ip <= 17.0:
        return 201 + (Ip - 10.1) * ((300-201)/(17.0-10.1))
    elif Ip <= 34.0:
        return 301 + (Ip - 17.1) * ((400-301)/(34.0-17.1))
    elif Ip >= 34.1:
        return 401 + (Ip - 34.1) * (598/34.1)
    else:
        return 0
    
df['CO_SubIndex'] = df['CO_AVG8H'].apply(lambda Ip: SubIndex_CO(Ip)).astype('int')
```

```python
def SubIndex_Ozone(Ip):
    if Ip <= 50:
        return 0 + (Ip - 0) * ((50 - 0) / (50 - 0))
    elif Ip <= 100:
        return 51 + (Ip - 51) * ((100-51) / (100-51))
    elif Ip <= 168:
        return 101 + (Ip - 101) * ((200-101) / (168-101))
    elif Ip <= 208:
        return 201 + (Ip - 169) * ((300-201)/(208-169))
    elif Ip <= 748:
        return 301 + (Ip - 209) * ((400-301)/(748-209))
    elif Ip >= 749:
        return 401 + (Ip - 749) * (598/749)
    else:
        return 0
    
df['Ozone_SubIndex'] = df['Ozone_AVG8H'].apply(lambda Ip: SubIndex_Ozone(Ip)).astype('int')
```

```python
def SubIndex_NO2(Ip):
    if Ip <= 40:
        return 0 + (Ip - 0) * ((50 - 0) / (40 - 0))
    elif Ip <= 80:
        return 51 + (Ip - 41) * ((100-51) / (80-41))
    elif Ip <= 180:
        return 101 + (Ip - 81) * ((200-101) / (180-81))
    elif Ip <= 280:
        return 201 + (Ip - 181) * ((300-201)/(280-181))
    elif Ip <= 400:
        return 301 + (Ip - 281) * ((400-301)/(400-281))
    elif Ip >= 401:
        return 401 + (Ip - 401) * (598/401)
    else:
        return 0
    
df['NO2_SubIndex'] = df['NO2_AVG24H'].apply(lambda Ip: SubIndex_NO2(Ip)).astype('int')
```

```python
def AQI_Category(Ip):
    if Ip <= 50:
        return "Good"
    elif Ip <= 100:
        return "Satisfactory"
    elif Ip <= 200:
        return "Moderate"
    elif Ip <= 300:
        return "Poor"
    elif Ip <= 400:
        return "Very Poor"
    elif Ip > 400:
        return "Severe"
    else:
        return np.NaN

df["Checks"] = (df["PM10_SubIndex"] > 0).astype(int) + \
                (df["PM2.5_SubIndex"] > 0).astype(int) + \
                (df["SO2_SubIndex"] > 0).astype(int) + \
                (df["CO_SubIndex"] > 0).astype(int) + \
                (df["NO2_SubIndex"] > 0).astype(int) + \
                (df["Ozone_SubIndex"] > 0).astype(int)

df["AQI_VALUE"] = round(df[["PM2.5_SubIndex", "PM10_SubIndex", "SO2_SubIndex",
                                "CO_SubIndex", "NO2_SubIndex", "Ozone_SubIndex"]].max(axis = 1)).astype('int')
df.loc[df["PM2.5_SubIndex"] + df["PM10_SubIndex"] <= 0, "AQI_VALUE"] = np.NaN
df.loc[df.Checks < 3, "AQI_VALUE"] = np.NaN


df["AQICategory"] = df["AQI_VALUE"].apply(lambda Ip: AQI_Category(Ip))
```
